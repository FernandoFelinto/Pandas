```python
import pandas as pd
```

```python
df = pd.read_excel("Dados.xlsx")
```

```python
# USANDO LOC
# 'df.loc[0]' = uma linha ou um intervalo 'df.loc[0:3]'
# Pode selecionar linhas específicas fora de intervalos df.loc[[2, 4, 7]]
# Seleciona um intervalo e pode mostrar os dados de uma coluna específica df.loc[2:5, "AddressLine1"]
# df.loc[3:6, ["AddressLine1", "City"]] -> duas colunas específicas
df.loc[0:3, "AddressLine1":"City"] #-> usa um intervalo de índices e um intervalo de colunas 
```

```python
# USANDO ILOC
# usa apenas os números
# no intervalo, 'iloc' não inclui o último número [3:5] -> 3, 4
# 'df.iloc[2:6, 1:4]' -> pega um intervalo de linhas [2:6] e um intervalo de colunas [1:4]
# df.iloc[[2,5], [True, False]]
df.iloc[[2,5], 1:4]
```
